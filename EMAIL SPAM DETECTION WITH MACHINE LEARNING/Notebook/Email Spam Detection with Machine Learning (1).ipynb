<span style="font-size: 14px; border: 2px solid #ccc;background-color: #ECF2FF; padding: 10px; border-radius: 10px; border: dashed; display: inline-block; margin-bottom: 8px;">

In the context of Natural Language Processing (NLP), feature engineering refers to the process of transforming raw text information into a format that can be used to train machine learning models. The goal of feature engineering is to create meaningful and pertinent features that capture what is most important in the text. This way, the model can learn to recognize patterns and make correct predictions. In NLP, raw text input tends to be unstructured, and it is often multidimensional. Extracting useful information from raw text can have a significant effect on the performance of an NLP model.

For the purpose of data exploration, I am creating new features in the dataset depicting the no. of character, words and sentences as follows:
    <li><span style="font-weight: bold;">No_of_Characters</span>: Number of characters in the text message</li>
    <li><span style="font-weight: bold;">No_of_Words</span>: Number of words in the text message</li>
    <li><span style="font-weight: bold;">No_of_Sentences</span>: Number of sentences in the text message</li>
</span>

<h3><span style="color:#ee4137;font-size:20px;">1. Text Length and Structure Analysis</span></h3>

```python
# Adding more columns to show the number of characters, words, and sentences in each email text with NLTK library  

data["No_of_Characters"] = data["Text"].apply(len)
data["No_of_Words"]=data.apply(lambda x: nltk.word_tokenize(x["Text"]), axis=1).apply(len)
data["No_of_Sentences"]=data.apply(lambda x: nltk.sent_tokenize(x["Text"]), axis=1).apply(len)
```

```python
# Defining a function to clean up the text
def Clean(Text):
    email = re.sub('[^a-zA-Z]', ' ', Text) # Replacing all non-alphabetic characters with a space
    email = email.lower() # converting to lowercase
    email = email.split() # splits the cleaned text sms into a list of words
    email = ' '.join(email) # joins the list of words back into a string, using a single space as the separator between the words
    return email

data["Clean_Text"] = data["Text"].apply(Clean)

# Lets have a look at a sample of texts after cleaning
print("\033[31;1;4m Sample of the First 5 Texts after cleaning:\033[0m",*data["Clean_Text"][:5], sep = "\n")
```

<span style="font-weight: bold;">Stopwords</span> are frequently occurring words(such as few, is, an, etc). These words hold meaning in sentence structure, but do not contribute much to language processing in NLP. For the purpose of removing redundancy in our processing, we will be removing those. NLTK library has a set of default stopwords that we will be removing.

```python
# Function to remove stopwords 
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    filtered_text = [word for word in text if word not in stop_words]
    return filtered_text

data["No_Stopword_Text"] = data["Tokenized_Text"].apply(remove_stopwords)

print("\033[31;1;4m Sample of the First 5 Texts after Removing the Stopwords:\033[0m",*data["No_Stopword_Text"][:5], sep = "\n")
```

<span style="font-weight: bold;">Stemming</span> is the process of getting the root form of a word. Stem or root is the part to which inflectional affixes are added. The stem of a word is created by removing the prefix or suffix of a word. It goes back to the etymology of the word. Languages evolve over time. Many different languages branch into each other; for example, English is a derivative of Latin. Thus, stemming a word takes it back to the root word.

<span style="font-weight: bold;">Lemmatization</span> also converts a word to its root form. However, the difference is that lemmatization ensures that the root word belongs to the language one is dealing with, in our case it is English. If we use lemmatization the output would be in English.
    
In a typical preprocessing pipeline, lemmatization is generally performed before stemming. The reason for this order is rooted in the goals and characteristics of these two text normalization techniques.
    
</span>

<h4><span style="color:#3887BE;font-size:20px;">Lemmatization on No Stopword Text</span></h4>

```python
lemmatizer = WordNetLemmatizer()

# lemmatize string
def lemmatize_word(text):
    lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in text]
    return lemmas

data["Lemmatized_Text"] = data["No_Stopword_Text"].apply(lemmatize_word)
print("\033[31;1;4m The First 5 Texts after Lemitization:\033[0m",*data["Lemmatized_Text"][:5], sep = "\n")
```

```python
# Importing required modules
from nltk.stem import PorterStemmer

# Creating the class object
stemmer = PorterStemmer()

#Stemming the words
def stemming_word(text):
    stem = [stemmer.stem(word) for word in text]
    return stem

data["Stemming_Text"] = data["No_Stopword_Text"].apply(stemming_word)
print("\033[31;1;4m The First 5 Texts after Stemming on No_Stopword_Text :\033[0m",*data["Stemming_Text"][:5], sep = "\n")
```

<span style="font-weight: bold;">TF-IDF</span> in NLP stands for Term Frequency ‚Äì Inverse document frequency. In NLP cleaned data needs to be converted into a numerical format where each word is represented by a matrix. This is also known as word embedding or Word vectorization.

The formula for TF-IDF for a term (word) "t" in a document "d" within a collection of documents is given by:

<span style="font-weight: bold;">TF-IDF(d,t)</span> = TF(d,t) √ó IDF(t) , where,

<li><span style="font-weight: bold;">TF(t,d)</span> represents the Term Frequency, which measures how often the term "t" occurs in the document "d." It is usually normalized to prevent bias towards longer documents.

<span style="font-weight: bold;">Term Frequency (TF)</span> = (Frequency of a term in the document) / (Total number of terms in documents)</li>

<li><span style="font-weight: bold;">IDF(t)</span> is the Inverse Document Frequency, which measures the rarity of the term "t" across all documents in the dataset. It is calculated as,
 
<span style="font-weight: bold;">Inverse Document Frequency(IDF)</span> = log( (total number of documents) / (number of documents with term t))</li>

I will be using TfidfVectorizer() to vectorize the preprocessed data.

<h3><span style="color:#ee4137;font-size:20px;">3. Model Training</span></h3>

```python
# Testing on the following classifiers
classifiers = [LogisticRegression(solver = 'liblinear', penalty = 'l1'),
               RandomForestClassifier(n_estimators = 50, random_state = 2),
               KNeighborsClassifier(), 
               SVC(kernel= "sigmoid", gamma  = 1.0),
               MultinomialNB(),
               BaggingClassifier(n_estimators = 50, random_state = 2),
               AdaBoostClassifier(n_estimators = 50, random_state = 2),
               GradientBoostingClassifier(n_estimators = 50, random_state = 2),
               XGBClassifier(n_estimators = 50, random_state = 2),
               DecisionTreeClassifier(max_depth = 5)]

for cls in classifiers:
    cls.fit(X_train, y_train)

# Dictionary of pipelines and model types for ease of reference
pipe_dict = {0: "Logistic", 1: "RandomForest", 2: "KNeighbours",3: "SVC", 4: "NaiveBayes", 5: "Bagging", 6: "AdaBoost",
             7: "GradientBoosting", 8: "XGBoost", 9: "DecisionTree"}
```

```python
# Cross Validation 
for i, model in enumerate(classifiers):
    cv_score = cross_val_score(model, X_train,y_train,scoring="accuracy", cv=10)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

<h2><center><p style="color: black;height: 40px; padding: 5px;background: #a7c5d4; text-align:center; border-radius: 15px 50px; font-size:25px;">üìà Model Evaluation </p></center></h2>

<span style="font-size: 14px; border: 2px solid #ccc;background-color: #ECF2FF; padding: 10px; border-radius: 10px; border: dashed; display: inline-block; margin-bottom: 8px;">

<span style="font-weight: bold;">Model Evaluation</span> entails evaluating the efficiency and performance of the developed models. In this section, we will test the models on Test set by using the following ways to assess a model's performance in model evaluation:

<li><span style="font-weight: bold;">Confusion Matrix:</span> To assess a classification model‚Äôs effectiveness, a confusion matrix is a tabular representation.</li>
<li><span style="font-weight: bold;">Accuracy Report:</span> Accuracy Report is a widely used metric that measures the overall correctness of a model‚Äôs predictions.</li>

<h3><span style="color:#ee4137;font-size:20px;">1. Evaluate all the Models</span></h3>

```python
# creating lists of various performance metrics
precision =[]
recall =[]
f1_score = []
trainset_accuracy = []
testset_accuracy = []

for i in classifiers:
    pred_train = i.predict(X_train)
    pred_test = i.predict(X_test)
    prec = metrics.precision_score(y_test, pred_test)
    recal = metrics.recall_score(y_test, pred_test)
    f1_s = metrics.f1_score(y_test, pred_test)
    train_accuracy = i.score(X_train,y_train)
    test_accuracy = i.score(X_test,y_test)
  
    # Appending scores
    precision.append(prec)
    recall.append(recal)
    f1_score.append(f1_s)
    trainset_accuracy.append(train_accuracy)
    testset_accuracy.append(test_accuracy)
```

```python
# initialise data of lists.
eval_data = {'Precision':precision, 'Recall':recall, 'F1score':f1_score, 'Accuracy on Trainset':trainset_accuracy,
             'Accuracy on Testset':testset_accuracy}

# Create pandas DataFrame with classifier names as indices.
classifier_model = [cls.__class__.__name__ for cls in classifiers]
Results = pd.DataFrame(eval_data, index=classifier_model)
```

```python
# Sort the DataFrame by the 'rank' column in ascending order
Results = Results.sort_values(by='Accuracy on Testset')
Results.style.background_gradient(cmap='PuBu')
```

<span style="font-size: 14px; border: 2px solid #ccc;background-color: #F5F5F5; padding: 10px; border-radius: 10px; display: inline-block; margin-bottom: 8px;">
<span style="font-weight: bold;">Observation:</span>The XGBoost classifier accurately identified the email messages as spam or not spam with highest accuracy of <span style="font-weight: bold;">98.2%</span> on test data and Precision of <span style="font-weight: bold;">95.5%</span>, followed by SVM model with <span style="font-weight: bold;">97.8%</span> accuracy on test data and Precision of <span style="font-weight: bold;">95.4%</span>
</span> 

<h3><span style="color:#ee4137;font-size:20px;">2. Rank the Performance Metrics</span></h3>

```python
# Create new columns for each criterion
Results['Precision_Rank'] = Results['Precision'].rank(ascending=False)
Results['Recall_Rank'] = Results['Recall'].rank(ascending=False)
Results['F1score_Rank'] = Results['F1score'].rank(ascending=False)
Results['Trainset_Accuracy_Rank'] = Results['Accuracy on Trainset'].rank(ascending=False)
Results['Testset_Accuracy_Rank'] = Results['Accuracy on Testset'].rank(ascending=False)

# Calculate the mean of the ranks to create an overall rank
Results['Overall_Rank'] = Results[['Precision_Rank', 'Recall_Rank', 'F1score_Rank', 'Trainset_Accuracy_Rank', 
                                   'Testset_Accuracy_Rank']].mean(axis=1)

# Create new column 'Rank' from the Overall_Rank
Results['Rank'] = Results['Overall_Rank'].rank(ascending=True, method='first').astype(int)

# Sort the DataFrame by the 'Rank' column in ascending order
Results = Results.sort_values(by='Rank')
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15,20))

for cls, ax in zip(classifiers, axes.flatten()):
    y_pred = cls.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", ax=ax, cmap=cmap2)
    ax.set_title(type(cls).__name__,fontsize=14, fontweight='bold')

plt.suptitle('Confusion Matrix of All the Models', size = 18, fontweight = 'bold', y=1)
plt.tight_layout(pad = 2.0)
plt.show()
```
